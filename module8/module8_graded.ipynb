Probability of Rain on Consecutive Days

You work as a junior data analyst in a city's weather forecasting department. The meteorological model outputs the probability of rain for each day based on atmospheric conditions, humidity, wind patterns, and historical data.


For every day in the near future, the system forecasts a 30% chance of rain


Write a Python function calculate_consecutive_rain_probability(n) that computes the probability that it will rain on all n consecutive days, assuming the events are independent.


Input Format


n: The number (int) of consecutive days

Output Format


The probability (float) that it will rain on all n consecutive days

Constraints


The computed probability is rounded to four decimal places
n ≥ 1

Example Case 1


Input


3

Output


0.027


Example Case 2


Input


5


Output


0.0024

```python
# Libraries (do not edit)
from ast import literal_eval

# Function to calculate consecutive rain probability
def calculate_consecutive_rain_probability(n):
    # Write your code here

    if n < 0:
        return 0

    probability_per_day = 0.3
    final = probability_per_day ** n

    return final

# Input and output processing (do not edit)
print(round(calculate_consecutive_rain_probability(literal_eval(input())), 4))
```

Joint Probability of Product Preferences

You are working as a data analyst for an e-commerce platform that collects feedback on product preferences. The platform shows users a variety of products and asks them to rate their likelihood of liking each one.


Based on historical data, the probability that a customer likes a particular product is estimated, and the results for each product are stored in a dictionary. The probabilities for each product are independent of each other.


Your task is to write a function calculate_joint_probability(products_and_probabilities) that returns the joint probability that a customer likes all the products


Input Format


A dictionary where:


The keys are product names (str)
The values are the probabilities (floats) representing the likelihood that a customer likes each product
The dictionary must contain at least one product (i.e., it cannot be empty)

Output Format


The joint probability (float) that the customer likes all the products

Constraints


The computed probability is rounded to 4 decimal places

Example Case 1


Input


{'headphones': 0.7, 'backpack': 0.5, 'laptop': 0.8}


Output


0.28


Example Case 2


Input


{'headphones': 1.0, 'backpack': 1.0, 'laptop': 1.0}


Output


1.0

```python
# Libraries (do not edit)
from ast import literal_eval

def calculate_joint_probability(products_and_probabilities):
    # Your code here

    joint_probability = 1
    for item in products_and_probabilities:
        joint_probability = joint_probability * products_and_probabilities[item]

    return joint_probability

# Input and output processing (do not edit)
print(round(calculate_joint_probability(literal_eval(input())), 4))
```

Estimate Property Claim Risk from Events

Instructions

As a risk analyst at an insurance company, you are tasked with evaluating the likelihood of various claims for properties, including fire, flood, and theft. Based on historical data, you are provided with the probabilities of each individual claim, as well as the likelihood of multiple claims occurring simultaneously, either in pairs or all three together.


Your task is to calculate the probability that a randomly selected policyholder will make at least one type of claim


Write a function claim_risk_exposure(p_fire, p_flood, p_theft, p_fire_flood, p_fire_theft, p_flood_theft, p_all_three) that accepts the probabilities as inputs and returns the probability of at least one claim


Input Format


A tuple containing, in sequence
p_fire (float): Probability of fire claim
p_flood (float): Probability of flood claim
p_theft (float): Probability of theft claim
p_fire_flood (float): Probability of both fire and flood
p_fire_theft (float): Probability of both fire and theft
p_flood_theft (float): Probability of both flood and theft
p_all_three (float): Probability of all three events occurring together

Output Format


The probability (float) that a property has at least one claim

Constraints


0 ≤ p_fire ≤ 1
0 ≤ p_flood ≤ 1
0 ≤ p_theft ≤ 1
0 ≤ p_fire_flood ≤ 1
0 ≤ p_fire_theft ≤ 1
0 ≤ p_flood_theft ≤ 1
0 ≤ p_all_three ≤ 1
The computed probability is rounded to two decimals

Example Case 1


Input


(0.3, 0.25, 0.2, 0.1, 0.08, 0.05, 0.03)


Output

0.55


Example Case 2


Input


(0.45, 0.2, 0.35, 0.15, 0.12, 0.09, 0.05)

Output

0.69

```python
# Libraries (do not edit)
from ast import literal_eval

def claim_risk_exposure(p_fire, p_flood, p_theft, p_fire_flood, p_fire_theft, p_flood_theft, p_all_three):
    # Code here

    final = (p_fire + p_flood + p_theft) - (p_fire_flood + p_fire_theft + p_flood_theft) + (p_all_three)
    return final

# Input and output processing (do not edit)
print(round(claim_risk_exposure(*literal_eval(input())), 2))
```

Exact Rain Probability in a Fortnight

You work as a junior data analyst in a city's weather forecasting department. The meteorological model outputs the probability of rain for each day based on atmospheric conditions, humidity, wind patterns, and historical data.


For every day in the near future, the system forecasts a 30% chance of rain. Write a Python function calculate_rain_probability_in_fortnight(n) that computes the probability that it will rain on exactly n days out of the next 14 days, assuming the events are independent.


Input Format


n: The number (int) of days (out of 14) that you want to calculate the probability of rainfall for

Output Format


The probability (float) that it will rain on exactly n days out of the next 14 days

Constraints


0 ≤ n ≤ 14
The computed probability is rounded to four decimal places

Example Case 1

Input

3


Output

0.1943


Example Case 2

Input

5


Output

0.1963

```python
# Libraries (do not edit)
from ast import literal_eval
import math

# Function to calculate rain probability in a fortnight
def calculate_rain_probability_in_fortnight(n):
    # Write your code here

    if n < 0 or n > 14:
        return 0
    else:
        p = 0.3
        N = 14

        final = math.comb(N, n) * (p ** n) * ((1 - p) ** (N - n))
        return final

# Input and output processing (do not edit)
print(round(calculate_rain_probability_in_fortnight(literal_eval(input())), 4))
```
