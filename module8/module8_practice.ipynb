Books List Genres Distribution

You are working for an online library system where each book is assigned a single, exclusive genre (i.e., no book belongs to more than one genre). The system maintains a catalog of books, and each genre has a specific percentage in the distribution.


Write a Python function calculate_genre_probability(selected_genres) that takes a list of genres as input and calculates the probability that a randomly selected book belongs to any of the genres in the list.


The percentage distribution of genres in the catalog is as follows:


Drama: 30%
Poetry: 25%
Science: 15%
History: 10%
Fantasy: 5%
Mystery: 5%
Romance: 4%
Thriller: 3%
Biography: 2%
Self-Help: 1%

Input format

A list of genres (str) representing the genres to be selected from the catalog

Output format

The total probability (float) of selecting a book from any of the input genres, rounded to four decimal places

Constraints

The genre names provided in the input list match the names in the predefined catalog
The input list is not empty
The computed probability is rounded to four decimals

Example Case 1

Input

['Drama', 'Science']


Output

0.45



Example Case  2

Input

['Drama', 'Poetry', 'Science', 'History', 'Fantasy', 'Mystery', 'Romance', 'Thriller', 'Biography', 'Self-Help']


Output

1.0

```python
def calculate_genre_probability(selected_genres):
    # Code here
    if not selected_genres:
        return "String is empty"

    temp = 0
    for genre in selected_genres:
        temp =  temp + genre_distribution[genre]
    return temp

# Input and output processing (do not edit)
from ast import literal_eval
genre_distribution = {'Drama': 0.30, 'Poetry': 0.25, 'Science': 0.15, 'History': 0.10, 'Fantasy': 0.05, 'Mystery': 0.05, 'Romance': 0.04, 'Thriller': 0.03, 'Biography': 0.02, 'Self-Help': 0.01}
print(round(calculate_genre_probability(literal_eval(input())), 4))
```

Student Pass Probability From List of Dicts

You are a data analyst at a high school that is evaluating student performance in three core subjects: Mathematics, Science, and English. The administration wants to understand overall pass coverage — i.e., what's the probability that a randomly selected student has passed at least one subject.


To solve this, you need to write a Python function probability_of_pass(selected_students) that takes a list of dictionaries as input. Each dictionary contains three keys, corresponding to the subjects, with boolean values indicating whether the student has passed or failed in each respective subject. The function should calculate the probability that a randomly selected student has passed at least one of the subjects (Mathematics, Science, or English).


Input format


A list of dictionaries representing the performance of students. Each dictionary contains:
Three keys: 'Mathematics', 'Science', and 'English'
The values (bool) for these keys, where True indicates a pass and False indicates a fail in the respective subject

Output format


The probability (float) that a randomly selected student has passed at least one subject

Constraints


The input list contains at least one student
The computed probability is rounded to four decimals

Example Case 1

Input

[{'Mathematics': True, 'Science': False, 'English': False}]


Output

1.0


Example Case 2

Input

[{'Mathematics': False, 'Science': False, 'English': False}]


Output

0.0

```python
# Function to compute passing probability of students
def probability_of_pass(selected_students):
    # Code here

    #count number of students
    student_count = len(selected_students)
    passed = 0

    #if list of selected students is empty then return 0
    if not selected_students:
        return 0

    for student in selected_students:
        if student['Mathematics'] or student['Science'] or student['English']:
            passed += 1

    final = passed / student_count

    return final


# Input and output processing (do not edit)
from ast import literal_eval
print(round(probability_of_pass(literal_eval(input())), 4))
```

Many-Sided Die and Many Rolls

Instructions

In a game simulation involving a die with n sides, you need to calculate the probability of achieving a specific sum after rolling the die a certain number of times. For example, if you roll a 6-sided die twice, you may want to calculate the probability of obtaining a sum of 7.


To solve this, you need to write a Python function probability_of_outcome(num_sides, rolls, target_sum) that takes three parameters: the number of sides on the die, the number of times the die will be rolled, and the target sum that you want to calculate the probability for across the rolls.


The function should return the probability of obtaining the target sum after rolling the die the specified number of times


Input format


A tuple whose elements, respectively, are

num_sides: the number (int) of sides on the die
rolls: the number (int) of die rolls
target_sum: the target sum (int) to calculate the probability for

Output format


A floating-point number (float) representing the probability of rolling the target sum with the given number of rolls

Constraints


The face in contact of the floor should be considered to contain the value to be summed (this is to avoid cases where there isn't a face pointing upwards, such as in the case of a 4-sided die)
Each side of the die has an equal probability of landing
The sides are each numbered in sequence (e.g., a 3 sided die will be numbered 1, 2, and 3 on its respective sides)
num_sides >= 4 to avoid the situation of impossible die structures in 3D
1 <= rolls <= 6
The computed probability is rounded to four decimals

Example Case 1

Input

(6, 2, 7)


Output

0.1667


Example Case 2

Input

(4, 4, 4)


Output

0.0039

```python
# Libraries (do not edit)
from ast import literal_eval

def probability_of_outcome(num_sides, rolls, target_sum):
    # Write your code here
    if target_sum < rolls or target_sum > rolls * num_sides:
        return 0.0

    # dp[i][j] = number of ways to get sum j with i rolls
    dp = [[0] * (target_sum + 1) for _ in range(rolls + 1)]
    dp[0][0] = 1  # 1 way to get sum 0 with 0 rolls

    for i in range(1, rolls + 1):
        for j in range(target_sum + 1):
            for face in range(1, num_sides + 1):
                if j >= face:
                    dp[i][j] += dp[i-1][j - face]

    total_outcomes = num_sides ** rolls
    favorable_outcomes = dp[rolls][target_sum]
    probability = favorable_outcomes / total_outcomes
    return probability


# Input and output processing (do not edit)
print(round(probability_of_outcome(*literal_eval(input())), 4))
```

Defect Probability in Manufacturing

Instructions

You are a quality assurance analyst working in a factory that manufactures electronic components, such as chips or connectors. Each day, the factory produces thousands of units, and your job is to monitor the defect rate to ensure quality standards are met.


At the end of the production run, a random sample is tested, and the number of defective units is recorded. Your task is to calculate the probability that a randomly selected item from the entire batch is not defective.


Input Format


A tuple containing, in sequence

An integer n representing the size of the sample
An integer d representing the number of defective units in the sample

Output Format


A float value representing the probability that a randomly selected unit is not defective

Constraints


n > 0
d >= 0
n >= d
The sample accurately represents the batch
The computed probability is rounded to 4 decimal places

Example Case 1


Input


(100, 27)


Output


0.73


Example Case 2


Input

(6, 2)

Output


0.6667

```python
# Function to calculate the probability that a unit is not defective
def calculate_probability(n, d):
    # Write your code here

    if n < 0 or d < 0 or n < d:
        return 0

    probability = 1 - d/n
    return probability


# Input and output processing (do not edit)
from ast import literal_eval
n, d = literal_eval(input())
print(round(calculate_probability(n, d), 4))
```

Finding Overlap Between Campaigns

You work as a marketing analyst for an e-commerce company. Your team recently launched two promotional campaigns:


Campaign A: a discount code for mobile users
Campaign B: a coupon for newsletter subscribers

Customers may qualify for either campaign or both. In this case, you're told that every customer is qualified for at least one campaign. Your task is to calculate the probability that a randomly selected customer qualified for both Campaign A and Campaign B.


Write a function campaign_overlap_probability(p_A, p_B) that accepts the probability of a customer qualifying for Campaign A and for Campaign B and returns the probability that a customer qualified for both campaigns


Input Format


A tuple containing the following, in sequence

p_A (float): the probability of qualifying for Campaign A
p_B (float): the probability of qualifying for Campaign B

Output Format


A floating-point number (float) representing the probability that a customer qualified for both campaigns

Constraints


0 ≤ p_A ≤ 1
0 ≤ p_B ≤ 1
The computed probability is rounded to two decimals

Example Case 1

Input

(0.6531, 0.5872)


Output

0.24

Example Case 2

Input

(0.72345, 0.68987)

Output

0.41

```python
# Libraries (do not edit)
from ast import literal_eval

def campaign_overlap_probability(p_A, p_B):
    # Your code here

    final = p_A + p_B
    return abs(1 - final)

# Input and output processing (do not edit)
print(round(campaign_overlap_probability(*literal_eval(input())), 2))
```

Land on Final Square

In a game of Snakes and Ladders, you are given a board with numbered squares. You are  squares away from the finish line, where  can be any non-negative integer. The goal is to land exactly on the final square.


You roll a 6-sided die and can move between 1 to 6 squares per roll. If you exceed the target square, you go back the remaining steps from the end.


For example, if you are 2 squares away from the end (given by X), and you roll 3, you will end up 1 square away from the end

i.e., the path would be (X-2) -> (X-1) -> X -> (X-1)

You are tasked with calculating the probability that you will land exactly on the final square in a given number of rolls, k.


Write a Python function probability_of_landing_on_final_square(n, k) that calculates the probability of landing exactly on the final square after a set number of rolls in a game. The function should take two inputs: the number of squares remaining before reaching the final square and the number of dice rolls. The output should be the probability that, after rolling the dice the specified number of times, you will land exactly on the final square.


Input format

A tuple containing, sequentially
n (int): represents how many squares are left before the final square
k (int): represents the number of rolls

Output format

A floating-point number (float) representing the probability that you will land exactly on the final square in k rolls

Constraints

n ≥ 1
k ≥ 1
The computed probability is rounded to four decimals

Example Case 1

Input

(4, 1)


Output

0.1667


Example Case  2

Input

(50, 1)


Output

0.0

```python
def probability_of_landing_on_final_square(n, k):
    # Write your code here
    total_outcomes = 6 ** k  # Total possible outcomes for k rolls of a 6-sided die
    valid_outcomes = 0

    # List to keep track of all possible positions after each roll
    positions = [n]  # Initial position is n squares away from the final square

    # Simulate k rolls
    for _ in range(k):
        next_positions = []

        # For each current position, simulate rolling each number (1 to 6)
        for pos in positions:
            for roll in range(1, 7):
                new_pos = pos - roll
                if new_pos < 0:
                    new_pos = abs(new_pos)  # Bounce-back if we exceed the target square
                next_positions.append(new_pos)

        # Update positions for the next roll
        positions = next_positions

    # Count how many valid outcomes land exactly on the final square (position 0)
    valid_outcomes = positions.count(0)

    # Calculate probability
    return valid_outcomes / total_outcomes

# Input and output processing (do not edit)
from ast import literal_eval
print(round(probability_of_landing_on_final_square(*literal_eval(input())), 4))
```
