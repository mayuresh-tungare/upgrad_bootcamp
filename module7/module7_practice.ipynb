Instructions

EduTech, an online education platform, is planning to launch a new study group program. The platform allows students to join collaborative learning groups for specific subjects. Each study group has a fixed number of students, denoted as k, who can participate from a total of n students. The task is to calculate how many unique study groups of size k can be formed from a total of n students, where the order in which students are selected does not matter (i.e., this is a combination problem, not a permutation).



Input Format



An integer n representing the total number of students
An integer k representing the number of students to be selected for each study group

Output Format


A single float value that represents the number of unique study groups

Constraints


1 ≤ k ≤ n ≤ 100

Example Case 1


Inputs


6

3



Output


20.0


Example Case 2


Inputs

5

2


Output

10.0

```python
# Taking the input
n=int(input())
k=int(input())

# Function to calculate factorial
def factorial(n):
    # Write your code here

    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)


# Function to calculate combinations (n choose k)
def calculate_combinations(n, k):
    # Write your code here

    ans = factorial(n) / (factorial (k) * factorial(n-k))
    return ans


# Print the output
print(calculate_combinations(n, k))
```

Category-wise Product Bundle Permutations

You work for an online retail store. For a limited-time promotional offer, the marketing team wants to create different product bundles. Each bundle should include one unit of k distinct products selected from the list of available items.


The list of products and their categories is stored in a dictionary, where:


Keys represent product names
Values represent their categories

You are tasked with writing a function called generate_bundle_permutations(products, k) that:


Generates all valid permutations of size k from the given products dictionary
Each permutation must be a tuple of k product names
All products in a tuple must come from different categories
The result must be lexicographically sorted

Note: In this solution, we use the permutations function from Python's itertools module


Syntax of itertools.permutations()

                  itertools.permutations(iterable, r)    
              

iterable: The collection (list, string, etc.) from which permutations are to be generated

r: The length of the permutation (i.e., how many elements should be in each tuple)


Input Format


A dictionary products where the keys represent product names (str) and the values represent their categories (str)
An integer k, representing the size of the product bundles to be selected

Output Format


A list of tuples, where each tuple contains k distinct product names
The products in each bundle must belong to different categories
The output should be lexicographically sorted

Constraints


N/A

Example Case 1


Inputs


{ 'pen': 'stationery', 'notebook': 'stationery', 'book': 'books', 'lamp': 'electronics' }

3



Output


[('book', 'lamp', 'notebook'), ('book', 'lamp', 'pen'), ('book', 'notebook', 'lamp'), ('book', 'pen', 'lamp'), ('lamp', 'book', 'notebook'), ('lamp', 'book', 'pen'), ('lamp', 'notebook', 'book'), ('lamp', 'pen', 'book'), ('notebook', 'book', 'lamp'), ('notebook', 'lamp', 'book'), ('pen', 'book', 'lamp'), ('pen', 'lamp', 'book')]



Example Case 2



Inputs


{ 'shampoo': 'toiletries', 'toothbrush': 'toiletries', 'chips': 'snacks' }

3

```python
# Import permutations to generate all possible ordered combinations of elements from the list
from itertools import permutations
from itertools import combinations
from itertools import product

# Taking the input
products = eval(input())
k = int(input())

# Function to generate product bundle permutations within constraints
def generate_bundle_permutations(products, k):
    # Write your code here

    # Group products by category
    category_groups = {}
    for prod, category in products.items():
        category_groups.setdefault(category, []).append(prod)

    # Get categories and check if we have enough
    categories = list(category_groups.keys())
    if len(categories) < k:
        return []

    # Generate all valid product combinations (one from each category)
    result = []
    for cat_combo in combinations(categories, k):
        # Get one product from each category in combo
        selections = [category_groups[cat] for cat in cat_combo]
        for prod_combo in product(*selections):
            # Add all permutations of this valid combination
            result.extend(permutations(prod_combo))

    return sorted(result)

# Print the output
print(generate_bundle_permutations(products, k))
```

Generate Product Combos Within Budget

In an e-commerce platform, managing and maximising customer purchases within a given budget is essential for sales optimisation. Retailers often offer multiple product options, but customers are constrained by their budget, so it’s important to identify combinations of products they can afford. By generating valid product combinations within the budget, businesses can help customers make informed purchasing decisions, increase sales, and enhance customer satisfaction by offering options that fit their financial limits.


Your task is to write a Python function generate_combinations_within_budget(products, budget) that:



Takes a dictionary of products (product name as key and price as value) and a budget

Generates all valid combinations of products where:
Each product is selected at most once
The total price of each combination does not exceed the budget
Only combinations containing two or more products are considered

Returns and prints the combinations in the same order they were generated


Input Format

A dictionary products, where each key is the product name (str) and the corresponding value is the product price (int)
Budget (int) representing the total available amount to spend on products

Output Format

Print each valid combination of product names that fits within the given budget should be printed on a separate line

If no valid combinations are found, print 'No valid combinations within the budget.'


Constraints

Each product price and the budget are positive integers

Example Case 1

Inputs


{'Milk': 30, 'Juice': 50, 'Bread': 20, 'Butter': 40}

100


Output

['Milk', 'Juice']['Milk', 'Juice', 'Bread']['Milk', 'Bread']['Milk', 'Bread', 'Butter']['Milk', 'Butter']['Juice', 'Bread']['Juice', 'Butter']['Bread', 'Butter']


Example Case 2


Inputs


{'Shirt': 100, 'Jeans': 200, 'Jacket': 500}

50


Output


No valid combinations within the budget.

```python
# Taking the input
products = eval(input())
budget = int(input())
from itertools import combinations

# Function to generate valid product combinations within budget (without using itertools)
def generate_combinations_within_budget(products, budget):
    # Write your code here

    product_names = list(products.keys())
    product_prices = list(products.values())

    valid_combinations = []

    # Helper function to generate combinations recursively
    def find_combinations(start, current_combination, current_sum):
        # Write your code here
        # If the current sum is within budget and has more than one item, add the combination
        if current_sum <= budget and len(current_combination) > 1:
            valid_combinations.append(current_combination)

        # Recursively add remaining products
        for i in range(start, len(product_names)):
            if current_sum + product_prices[i] <= budget:
                find_combinations(i + 1, current_combination + [product_names[i]], current_sum + product_prices[i])

    # Start the recursive process
    find_combinations(0, [], 0)

    return valid_combinations

# Print the output
valid_combos = generate_combinations_within_budget(products, budget)

if valid_combos:
    for combo in valid_combos:
        print(combo)
else:
    print('No valid combinations within the budget.')
```

Job Role Predictor

Instructions

You are a data analyst at Career Craft, a platform that helps users identify suitable job roles based on their self-assessed skill ratings. Users rate themselves on a scale of 1 to 5 in the given skills, namely 'Communication', 'Problem Solving', 'Technical Knowledge', 'Creativity', and 'Leadership'. Your task is to analyse these ratings and predict the most suitable job role from a predefined set of roles, namely 'Software Developer', 'Product Manager', 'UX Designer', 'Data Analyst', and 'Sales Executive'.


Implement a Python function predict_job_role(ratings) that accepts a list of skill ratings and returns the most suitable job role based on a rule-based logic.


Rules


Software Developer: High Technical Knowledge (≥ 4), High Problem Solving (≥ 4), Low Communication (≤ 3)
Product Manager: High Communication (≥ 4), High Leadership (≥ 4), Moderate to High Problem Solving (≥ 3)
UX Designer: High Creativity (≥ 4), Moderate to High Communication (≥ 3), Low Technical Knowledge (≤ 3)
Data Analyst: High Technical Knowledge (≥ 4), High Problem Solving (≥ 4), Low Leadership (≤ 3)
Sales Executive: High Communication (≥ 4), Moderate to High Leadership (≥ 3), Moderate to High Creativity (≥ 3)
If none of the above conditions are met, default to 'Product Manager'

Input Format


A tuple of 5 integers

Output Format


A string representing the predicted job role

Constraints


Ratings must be integers between 1 and 5
All five skill categories must be present
No value should be missing, empty, or zero
Only the five predefined job roles are allowed
Each integer in the input should represent a rating (typically from 1 to 5) for a specific skill in the order 'Communication', 'Problem Solving', 'Technical Knowledge', 'Creativity', 'Leadership'

Example Case 1


Input


3, 5, 5, 3, 2


Output


Software Developer


Example Case 2


Input


5, 4, 3, 4, 5


Output


Product Manager

```python
# Import pandas library as 'pd' for data manipulation and analysis
import pandas as pd

# Taking the input (do not edit)
user_input = input()

# Convert input string to list of integers
values = [int(x.strip()) for x in user_input.split(',')]

# Write your code here
# Define skill names in order
# Create a dictionary mapping skills to their ratings


# Define the prediction function
def predict_job_role(values):

    # validate if all values are present
    if len(values) != 5:
        return "Not all ratings available"

    if not all(values):
        return "Atleast one of the ratings is missing, empty or zero"

    # Extract ratings from values list
    comm = values[0]
    prob_solve = values[1]
    tech = values[2]
    creat = values[3]
    lead = values[4]

    # Software Developer
    if tech >= 4 and prob_solve >= 4 and comm <= 3:
        return 'Software Developer'

    # Product Manager
    if comm >= 4 and lead >= 4 and prob_solve >= 3:
        return 'Product Manager'

    # UX Designer
    if creat >= 4 and comm >= 3 and tech <= 3:
        return 'UX Designer'

    # Data Analyst
    if tech >= 4 and prob_solve >= 4 and lead <= 3:
        return 'Data Analyst'

    # Sales Executive
    if comm >= 4 and lead >= 3 and creat >= 3:
        return 'Sales Executive'

    # Default
    return 'Product Manager'

# Print the output for the predicted job role (do not edit)
print(predict_job_role(values))
```

Delivery Status Tracker

You are working with a logistics platform that tracks the status of deliveries in various cities. The system logs each delivery with an order ID, city, and status, and your task is to analyse this data to provide summary statistics.


You need to count the number of orders in each status category, list the orders that are in transit or cancelled, and identify the cities where deliveries were successful.


You are given a list of such tuples. Your task is to:


Count how many orders are in each status category
Create a list of all order_id that are currently 'in transit', or 'cancelled', sorted alphabetically
Return a set of all unique cities where at least one delivery was successfully 'delivered'

Input Format

A list of tuples, each of the form (order_id, city, status)
order_id (str)
city (str)
status (str): one of 'delivered', 'in transit', or 'cancelled'

Output Format

Return a list with three elements:

A dictionary with keys 'delivered', 'in transit', and 'cancelled' and their respective counts
A sorted list of order_id with status 'in transit' or 'cancelled'
A lexicographically sorted list of cities where at least one order was 'delivered'. Lexicographical order is essentially alphabetical order, but it applies to any sequence of characters.

Constraints

order_id and city are non-empty strings
status is always one of: 'delivered', 'in transit', 'cancelled'
There may be 0 or more delivery records

Example Case 1


Input


[('OD101', 'Mumbai', 'delivered'), ('OD102', 'Delhi', 'in transit'), ('OD103', 'Mumbai', 'cancelled'), ('OD104', 'Bangalore', 'delivered'), ('OD105', 'Delhi', 'in transit')]


Output

[{'delivered': 2, 'in transit': 2, 'cancelled': 1}, ['OD102', 'OD103', 'OD105'], ['Bangalore', 'Mumbai']]



Example Case 2

Input


[]


Output


[{'delivered': 0, 'in transit': 0, 'cancelled': 0}, [], []]

```python
# Import literal_eval to safely evaluate string input as a Python literal
from ast import literal_eval

# Taking the input
records = literal_eval(input())
order_ids = []
cities = []
order_status = {'delivered':0,'in transit':0,'cancelled':0}

# Write your code here
for record in records:

    od_id, city, od_st = record
    #print(od_id, city, od_st)

    order_status[od_st] += 1

    if od_st == 'in transit' or od_st == 'cancelled':
        order_ids.append(od_id)

    if city not in cities and od_st == 'delivered':
        cities.append(city)

final_list = [order_status,order_ids,sorted(cities)]

print(final_list)

# Print the output
#print([status_count, problematic_orders, delivered_cities])
```
