Write a recursive function to calculate the sum of the digits of a given number.



Input Format:

 An integer n



Output Format:

An integer representing the sum of the digits of n



Constraints:

0 <= n <= 100000





Sample Input 1:

123



Sample Output 1:

6



Sample Input 2:

1234



Sample Output 2:

10



```python
# Taking input
n = int(input())

def sum_of_digits(n):
    # Write your code
    if n < 10:
        return n
    return (n % 10 + sum_of_digits(n//10))

# Print the output
print(sum_of_digits(n))
```

Check Armstrong Number
Write a function that checks whether a number is an Armstrong number. A number is Armstrong if the sum of the cubes of its digits is equal to the number itself.



Input Format:

An integer num



Output Format:

Yes if the number is Armstrong, else No



Constraints:

0 <= num <= 9999



Sample Input:

153



Sample Output:

Yes



```python
#Taking input
num = int(input())

def is_armstrong(num):
    # Write your code
    total = 0
    n = num

    if n <= 0 or n >= 9999:
        return "Invalid"
    elif n >= 1 and n < 10:
        return "Yes"
    else:
        while n >= 1:
            total = total + (n % 10) ** 3
            n = n // 10

    if total == num:
        return "Yes"
    else:
        return "No"

# Print the output
print(is_armstrong(num))
```

Recursive Digit Product

Write a Recursive Function 'digit_product(n)' that returns the product of all digits in the number n.



Input Format:

An integer n



Output Format:

An integer representing the product of digits



Constraints:

0 <= n <= 100000
Assume n will never contain a 0 digit for now




Sample Input:

234



Sample Output:

24



```python
# Taking input
n = int(input())

def digit_product(n):
    # Write your code
    num = n

    if n > 0 and n <= 10000:
        if n > 10:
            return (n % 10) + digit_product (n // 10)
        else:
            return n

# Print the output
print(digit_product(n))
```

The second row will contain the next two odd numbers (the 2nd and 3rd in the overall sequence).
The third row will contain the next three odd numbers (the 4th, 5th, and 6th in the overall sequence), and so on.
In general, row i (from 1 to N) should contain i consecutive odd numbers from the ongoing sequence. The numbers within each row should be printed in a right-aligned manner.


Input Format:

A single integer N (1 ≤ N ≤ 15), representing the number of rows.



Output Format:

Output the pattern consisting of N rows.
 Each row should display i consecutive odd numbers, separated by a single space, with each row printed on a new line.

Find the Nth Term of an Arithmetic Progression
Write a function 'find_nth_arithmetic_term(a, d, n_term)' that finds the 'n_term-th' value in the arithmetic progression.



In an arithmetic progression (AP), each term after the first is obtained by adding a constant value called the common difference, d, to the previous term.



The first term is denoted by a, and the 'n_term-th' term can be computed using the formula:

T(n)=a+(n_term−1)×d

Where:

a: the first term of the progression.
d: the common difference between consecutive terms.
'n_term': the position of the term you want to find (e.g., 1st, 2nd, 3rd term).


Input Format:

Three inputs, each on a separate line:

a: Integer (no restrictions specified on magnitude).
d: Integer (no restrictions specified on magnitude).
n_term: Integer, where 1 ≤ n_term ≤ 100


Output Format:

An integer representing the 'n_term-th' value in the arithmetic progression.



Constraints:

The first term a (integer).
The common difference d (integer).
The term number 'n_term' (positive integer, 1 ≤ n_term ≤ 100).




Sample Input:

3

5

4



Sample Output:

18

```python
# Taking input
a = int(input())
d = int(input())
n_term = int(input())

def find_nth_arithmetic_term(a, d, n_term):

    # Write Your code
    if n_term >= 1 and n_term <= 100:
        T = a + (n_term - 1) * d
        return T

# Print the result
print(find_nth_arithmetic_term(a, d, n_term))
```

Check for Perfect Number

Write a function 'check_perfect_number(n)' that checks whether a given number n is a perfect number.

A perfect number is a positive integer that is equal to the sum of its proper divisors (excluding the number itself).



Input Format:

A single integer n



Output Format:

Return True if the number is perfect, otherwise return False.



Constraints:

n is a positive integer (1≤n≤106)





Sample Input:

6



Sample Output:

True

```python
# Taking input
n = int(input())

def check_perfect_number(n):

    # Write your code
    upper_limit = 10 ** 6
    total = 0
    if n >= 1 and n <= upper_limit:
        for i in range(1,n):
            if n % i == 0:
                total = total + i
        if total == n:
            return ("True")
        else:
            return ("False")
# Print the result
print(check_perfect_number(n))
```
