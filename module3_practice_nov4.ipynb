Write a program to find  Greatest Common Divisor (GCD) of Two Positive Numbers.



The Greatest Common Divisor (GCD) of two positive integers is the largest positive number that divides both numbers exactly without leaving a remainder.




For example, the GCD of 48 and 18 is 6 because 6 is the biggest number that divides both 48 and 18 completely.



Input Format: Two integers given on separate lines.

Output Format: Single integer representing the GCD of the two numbers.




Sample Input:

48

18




Sample Output:

6

```python
# Taking input
num1 = int(input())
num2 = int(input())

def gcd(a, b):
    # Write your code here
    highest_divisor = 1

    #determine the smaller number between num1 and num2
    if a <= b:
        lower_number = a
        higher_number = b
    else:
        lower_number = b
        higher_number = a

    #from lower_number to 1, keep looking for the highest divisor
    for i in range(lower_number, 1, -1):
        #check highest divisor of the smaller number (range) which also divides the higher number
        if (lower_number % i == 0) and (higher_number % i == 0):
            highest_divisor = i
            break

    #return the highest divisor
    return(highest_divisor)

# Print the output
print(gcd(num1, num2))
```

Write a program that checks if a given number is prime or not.

A prime number is a positive integer greater than 1 that has no divisors other than 1 and itself.



Constraints: The input number should be a positive integer greater than 1.




 Input Format:

Single integer n.




Output Format:

"Prime" if the number is prime, otherwise "Not Prime".



Sample Input:

7

Sample Output:

Prime

```python
# Taking input
num = int(input())

def is_prime(n):
    # Write your code

    #define prime and not_prime strings
    prime="Prime"
    not_prime="Not Prime"
    flag = 0

    #is number zero or negative?
    if n <= 0:
        #when number is zero or negative
        return not_prime
    else:
        #when number is postive, check for a divisor from 2 to n-1 (1 and n are divisors anyways)
        for i in range (2, n):
            if (n % i == 0):
                flag = 1 #setting flag to true if a divisor is found
                break
            else:
                continue

    #check for flag and return finding accordingly
    if flag:
        return not_prime
    else:
        return prime

# Print the output
print(is_prime(num))
```

Write a program that takes three side lengths as input and determines whether they form an equilateral, isosceles, or scalene triangle. Additionally, the program should verify whether the given side lengths form a valid triangle.



 Input Format:

Three positive integers are given on separate lines, representing side lengths of a triangle.



Output Format:

Print a single word on a new line:

"Equilateral"
"Isosceles"
"Scalene"
"Not a triangle"


Constraints:

The side lengths a, b, and c are positive integers.
Each side length satisfies 1 ≤ a, b, c ≤ 109
All inputs will be valid integers within the specified range.


Sample Input:

3

3

3

Sample Output:

Equilateral



```python
# Taking input
side1 = int(input())
side2 = int(input())
side3 = int(input())

def classify_triangle(a, b, c):
    # Write your code

    # define variables
    equilateral = "Equilateral"
    isosceles = "Isosceles"
    scalene = "Scalene"
    not_triangle = "Not a triangle"
    invalid = "Invalid input"

    # invalid input if any of sides is not a postive integer
    if (a<=0) or (b<=0) or (c<=0):
        return invalid

    # invalid input if any of sides is larger than 10**9
    if (a >= 10**9) or (b >= 10**9) or (c >= 10**9):
        return invalid

    # check if it is a triangle
    if (a+b>c) and (b+c>a) and (c+a>b):
        # check if it is an equilateral triangle
        if (a == b == c):
            return equilateral
        # check if it is an isoceles triangle
        if (a == b) or (b == c) or (c == a):
            return isosceles
        # check if it is a scalene triangle
        if (a != b) and (b != c) and (c != a):
            return scalene
    else:
        return not_triangle

# Print the output
print(classify_triangle(side1, side2, side3))
```

Write a function 'traffic_light_action(color)' that simulates a simple traffic light system with three colors: 'red', 'yellow', and 'green'.

The function should print the corresponding action for each color:

 "red": Print "Stop"
"yellow": Print "Prepare to stop"
"green": Print "Go"
If the input color is invalid, print "Invalid color".


Input Format:

A single string `color` that represents the traffic light color. The color must be one of the following: "red", "yellow", or "green".

Output Format:

The function should print the corresponding action for the color provided.



Constraints:

The input `color` is a string and must be one of the following format: "red", "yellow", or "green", and in lowercase.



Sample Input:

green

Sample Output:

Go

```python
# Taking input
color = input()

def traffic_light_action(color):
    # Write your code

    # define variables
    red = "Stop"
    yellow = "Prepare to stop"
    green = "Go"
    invalid = "Invalid color"

    match color:
        case "red":
            print(red)
        case "yellow":
            print(yellow)
        case "green":
            print(green)
        case _:
            print(invalid)

# Print the output
traffic_light_action(color)
```

Reverse a Number Without Converting to String

Write a program to reverse a Non-Negative Integer without converting it to a String.

The program should take an Integer and return the Integer Reversed.



Input Format:

Non-Negative Integer.



Output Format:

The Reversed Integer.



Constraints:

The input is a Non-Negative Intege.



Sample Input:

12345

Sample Output:

54321

```python
# Taking input
num = int(input())

def reverse_number(n):
    # Write your code
    invalid = "Invalid input"
    original = n
    reverse = 0
    digit = 0

    if original < 0:
        return invalid
    else:
        while original > 0:
            digit = original % 10
            reverse = digit + (reverse * 10)
            original = original // 10
        return reverse

# Print the output
print(reverse_number(num))
```

Difference of Sum of Squares and Square of Sum

Write a program that calculates the difference between the sum of the squares and the square of the sum for the first n natural numbers.


Specifically, given a positive integer n, compute:

The sum of the squares: 1² + 2² + ... + n²
The square of the sum: (1 + 2 + ... + n)²
Return the difference: (square of the sum) - (sum of the squares).


Input Format:

A single integer n representing the number of natural numbers.


Output Format:

An integer representing the difference between the square of the sum and the sum of the squares of the first n natural numbers.


Constraints:

1 <=n <=106


Sample Input:

10


Sample Output:

2640

```python
# Taking input
n = int(input())

def difference_sum_squares(n):
    # Write your code here

    #initialize variables
    sum_of_squares = 0
    square_of_sum = 0
    difference = 0

    # operations within loop from 0 to n+1
    for i in range(0,n+1):
        sum_of_squares = sum_of_squares + i**2
        square_of_sum = square_of_sum + i

    # operations outside of the loop
    square_of_sum = (square_of_sum ** 2)
    difference = (square_of_sum - sum_of_squares)
    return difference

# Print the output
print(difference_sum_squares(n))
```

**Sum of Digits Until Single Digit**

Write a Python function 'sum_until_single_digit(num)' that takes a Non-Negative Integer number and repeatedly sums its digits until the result is a single digit. Return that single-digit sum.


For example, given number = 9875,
The sum of digits is 9+8+7+5 = 29,

then sum digits of 29 → 2+9 = 11, and

Finally, sum the digits of 11 → 1+1 = 2.

The function returns 2.



Input Format:

A single non-negative integer num.



Output Format:

Return the single-digit integer obtained by repeatedly summing the digits of the number.



Constraints:

0 <= n<= 109



Sample Input:

9875

Sample Output:

2

```python
# Taking input
num = int(input())

def sum_until_single_digit(num):

    # Write your code
    # define variables
    n = num
    sum_of_digits = 0

    # negative number test
    while n > 0:
        sum_of_digits = sum_of_digits + (n % 10)
        n = n // 10

    if sum_of_digits < 10:
        return sum_of_digits
    else:
        return sum_until_single_digit(sum_of_digits)

# Print the output
print(sum_until_single_digit(num))
```

Write a Python function print_zigzag(n) that prints numbers from 1 to n in a zigzag pattern across three rows.

The zigzag pattern follows this repeating cycle every 4 numbers:

1st number in row 1
2nd number in row 2
3rd number in row 3
4th number in row 2


Each number is followed by a space.

Positions without a number should be filled with two spaces "  " to maintain alignment.



Input Format:

A single integer n representing the total number to print in the zigzag pattern.



Output Format:

Print 3 lines, each containing the numbers arranged in the zigzag pattern separated by spaces, with appropriate spacing for alignment.



Constraints:

1 <= n <=11



Sample Input:

10



Sample Output:

1       5       9    

  2   4   6   8   10    

     3       7      

```python
num = int(input())

def print_zigzag(n):
    row_map = [0, 1, 2, 1]  # maps index to row in cycle
    for r in range(3):  # for each row
        s = ""  # will build the string line by line
        for i in range(n):  # for each column (number 1 to n)
            if r == row_map[i % 4]:
                s += f"{i+1} "
            else:
                s += "  "
        print(s)

# Print the output
print_zigzag(num)
```

Convert Decimal Number to Binary String

Write a Python function 'decimal_to_binary(num)' that converts a Non-Negative Decimal Integer number to its Binary representation as a String.


Input Format:

A single non-negative integer number.


Output Format:

Return a string representing the Binary equivalent of the input decimal number.


Constraints:

0 <= n <= 109


Sample Input:

10


Sample Output:

1010  

```python
# Taking input
decimal_num = int(input())

def decimal_to_binary(num):
    # Write your code here

    #initiate binary string to empty
    binary = ""
    bit = 0

    if num <= 0:
        return 0
    else:
        while num > 0:
            bit = (num % 2)
            num = num // 2
            binary = binary + str(bit)
        binary = binary[::-1]

    return binary

# Print the output
print(decimal_to_binary(decimal_num))
```

Check if a Number is a Strong Number
Write a program to check whether a given number is a Strong Number or not.



A Strong Number is a number whose sum of the factorial of each digit is equal to the number itself.



For example,

145 is a strong number because: 1! + 4! + 5! = 1 + 24 + 120 = 145.



Input Format:

A single integer number.



Output Format:

Print True if the number is a Strong Number; otherwise, print False.



Sample Input:

145



Sample Output:

True

```python
# Taking input
n=int(input())

def factorial(n):
   # Write Your Code here
   if n == 0:
      return 1
   else:
      return n * factorial(n-1)

def is_strong_number(number):
   # Write Your Code here
   sum = 0
   n = number

   while n >= 1:
      digit = n % 10
      sum = sum + factorial(digit)
      n = n // 10

   if sum == number and number!=0:
      return "True"
   else:
      return "False"

# Print the output
print(is_strong_number(n))
```
