Count Vowels and Consonants in a String

Write a program that takes a string as input and counts the number of vowels and consonants in it.

Vowels include
 'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'.
Consonants are all other alphabetic letters (i.e., all letters except vowels).
All non-alphabetic characters (e.g., spaces, digits, punctuation) in the input string should be disregarded.


Input Format:

A string 's' consists of any valid ASCII character.


Output Format:

Count of vowels in the first line.
Count of consonants in the second line.


Constraints:

1 ≤ len(s) ≤ 105
The string may contain uppercase and lowercase characters, spaces, and punctuation, but only alphabetic characters should be counted.




Sample Input 1:

hello



Sample Output 1:

2

3



Sample Input 2:

Hi,Upgrad...!


Sample Output 2:

3

5

```python
# Taking input
s = input()

def count_vowels_and_consonants(s):
    # Write your code here

    if len(s) < 1 or len(s) >= 10**5:
        return False

    v_count = 0
    c_count = 0

    for char in s:
            if char in "aeiouAEIOU":
                v_count = v_count + 1
            else:
                if char.isalpha():
                    c_count = c_count + 1

    return (v_count,c_count)


# Print the Output
vowels, consonants = count_vowels_and_consonants(s)
print(vowels)
print(consonants)
```

Find Common Elements in Three Sorted Lists

Write a program that accepts three sorted lists of integers as input and returns a list containing the elements that are common to all three lists.



The program should be implemented efficiently, taking advantage of the fact that the input lists are already sorted. Avoid using unnecessary nested loops, and do not use sets or any external libraries that are not part of the standard Python curriculum up to this point.



If there are no common elements among the three lists, the program should output "None" (without quotes).



Input Format:

The program accepts three sorted lists as input, one list per line.


Output Format:

If common elements exist, output them as space-separated integers in a single line.
If no common elements exist, output "None" (without quotes).


Constraints:

Length of each list ≤ 10⁴.
Each list is sorted in non-decreasing order.
Each number is in the range: −10⁵ ≤ number ≤ 10⁵.
Lists may contain duplicates
No Nested loops




Sample Input:

[1, 5, 10, 20, 40, 80]

[6, 7, 20, 80, 100]  

[3, 4, 15, 20, 30, 70, 80, 120]



Sample Output:

20 80

```python
# Import the literal_eval library to evaluate string input as a Python safely
from ast import literal_eval

# Taking input
list1 = literal_eval(input())
list2 = literal_eval(input())
list3 = literal_eval(input())

# Efficient linear traversal using three pointers
def find_common_elements(l1, l2, l3):
    # Write your code here

    i, j, k = 0, 0, 0
    result = []
    len1, len2, len3 = len(l1), len(l2), len(l3)
    while i < len1 and j < len2 and k < len3:
        # Get current values
        a, b, c = l1[i], l2[j], l3[k]
        # Check if all three values are equal
        if a == b == c:
            # To avoid duplicates in the result
            if not result or result[-1] != a:
                result.append(a)
            i += 1
            j += 1
            k += 1
        else:
            # Find the smallest value, increment that pointer
            min_val = min(a, b, c)
            if a == min_val:
                i += 1
            if b == min_val:
                j += 1
            if c == min_val:
                k += 1
    return result if result else None

    # Input processing
    l1 = list(map(int, input().strip().strip('[]').split(',')))
    l2 = list(map(int, input().strip().strip('[]').split(',')))
    l3 = list(map(int, input().strip().strip('[]').split(',')))

    # Find common elements
    common = find_common_elements(l1, l2, l3)

    # Output result
    if common:
        print(' '.join(map(str, common)))
    else:
        print("None")

common_elements = find_common_elements(list1, list2, list3)

# Print the output
print(" ".join(map(str, common_elements)) if common_elements else "None")
```

Find Most Frequent Vowel and Consonants

Write a program that takes a string 's' consisting of only lowercase English letters (from 'a' to 'z') and performs the following operations:

Identify the vowel (one of 'a', 'e', 'i', 'o', or 'u')  that appears with the highest frequency in the string.
Identify the consonant (i.e., all other lowercase letters except vowels) that appears with the highest frequency.
Compute and return the sum of the frequencies of the most frequent vowel and the most frequent consonant.


Note: If multiple vowels or consonants have the same maximum frequency, you may choose any one of them.



The frequency of a letter 'x' is the number of times it occurs in the string.



Input Format:

A string 's' consisting of lowercase English letters only.



Output Format:

Return the sum of the frequencies of the most frequent vowel and consonant.



Constraints:

1 <= len(s) <= 100
's' consists of lowercase English letters only.




Sample Input 1:

"aeiaeia"



Sample Output 1:

3



Sample Input 2:

"successes"



Sample Output 2:

6

```python
# Taking input
s = input()

def max_vowel_consonant_sum(s):

    # Write your code here
    vowels = {'a', 'e', 'i', 'o', 'u'}
    freq = {ch: 0 for ch in s}
    for ch in s:
        freq[ch] += 1

    max_vowel_freq = 0
    max_consonant_freq = 0

    for ch in freq:
        if ch in vowels:
            max_vowel_freq = max(max_vowel_freq, freq[ch])
        else:
            max_consonant_freq = max(max_consonant_freq, freq[ch])

    return max_vowel_freq + max_consonant_freq

# Print the output
print(max_vowel_consonant_sum(s))
```

Merge Alternating Characters from Two Strings

Write a program that takes two strings, 'a' and 'b', as input and merges them by alternating characters from both strings.



The program's task is to construct and return a new string formed by alternating characters from each string.

Start by taking one character from string 'a', then one from string 'b', and continue this pattern.
If one string runs out of characters before the other, append the remaining part of the longer string to the result.


Input Format:

Two strings 'a'  and 'b' (can be of different lengths).
Each string contains only alphanumeric characters (a–z, A–Z, 0–9).
Strings should not be empty.


Output Format:

A single string formed by alternating characters from strings 'a' and 'b'.



Constraints:

0 ≤ len(a), len(b) ≤ 1000
Characters in strings 'a' and 'b' can be lowercase, uppercase, or digits.
Case-sensitive merge (i.e., 'A' ≠ 'a')




Sample Input:

abc

1234



Sample Output:

a1b2c34

```python
# Taking input
a = input()
b = input()

def merge_alternating(a, b):

    a = a.strip()
    b = b.strip()

    # Write your code here
    # perform basic validations
    if a.isalnum() and b.isalnum():
        if len(a) >= 0 and len(a) <= 1000:
            if len(b) >= 0 and len(b) <= 1000:

                #initiate new list
                result = []
                len_a, len_b = len(a), len(b)
                min_len = min(len_a, len_b)

                # Take one character from each string alternately up to the length of the shorter string
                for i in range(min_len):
                    result.append(a[i])
                    result.append(b[i])

                # Append the remaining characters from the longer string
                if len_a > len_b:
                    result.append(a[min_len:])
                elif len_a < len_b:
                    result.append(b[min_len:])

                return ''.join(result)

# Print the output
print(merge_alternating(a, b))
```

Partition List by Even and Odd

Write a program to rearrange the list such that all even numbers appear before all odd numbers. The relative order of even and odd numbers should be preserved as in the original list.



Input Format:

An integer 'n', the number of elements in the list.
A list of 'n' integers


Output Format:

A single list where all even numbers precede odd numbers, in their original order.



Constraints:

1 ≤ n ≤ 10⁴
-10⁴ ≤ arr[i] ≤ 10⁴


Sample Input:

6

1 2 3 4 5 6



Sample Output:

[2, 4, 6, 1, 3, 5]


```python
# Import the literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
n = int(input())
arr = literal_eval(input())

def partition_even_odd(arr):
    # Write your code here

    #perform basic validation
    if n < 1 or n >= 10**4:
        return "Invalid input"

    #initiate newlist
    evenlist = []
    oddlist = []

    for num in arr:
        if num % 2 == 0:
            evenlist.append(num)
        else:
            oddlist.append(num)

    return evenlist + oddlist

# Print the output
print(partition_even_odd(arr))
```

All Indices of Target Element

Write a function that returns a list of all indices where the target element occurs in the list.



The program should return the indices in ascending order. If the target does not appear in the list, return an empty list.



Input Format:

An integer n, representing the number of elements in the list.
A list of n integers.
An integer target, the element to find.


Output Format:

A list of integers representing the indices (0-based) where the target occurs in the list.



Constraints:

1 ≤ n ≤ 10⁴
-10⁴ ≤ lst[i], target ≤ 10⁴




Sample Input:

6

[1, 2, 3, 2, 4, 2]

2



Sample Output:

[1, 3, 5]

```python
# Import the literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
n = int(input())
lst = literal_eval(input())
target = int(input())


def find_all_indices(lst, target):
    # Write your code here

    #perform basic validation
    if n < 1 or n >= 10**4:
        return "Invalid input"

    #initiate output list
    output_list = []

    for i in range(0,n):
        if lst[i] == target:
            output_list.append(i)

    return output_list

# Print the output
print(find_all_indices(lst, target))
```

Product of All Elements Except Self

Write a function that takes a list of integers and returns a new list such that each element at index 'i' of the new list is the product of all the elements in the original list except the one at 'i'.

Avoid using division to handle cases involving zeros.
If the list contains only one element, return [1].


Input Format:

An integer 'n', the number of elements in the list.
A list of 'n' integers.


Output Format:

A list of integers representing the product of all elements except self for each position.



Constraints:

1 ≤ n ≤ 10³
-100 ≤ arr[i] ≤ 100




Sample Input 1:

4

[1, 2, 3, 4]



Sample Output 1:

[24, 12, 8, 6]



Sample Input 2:

1

[7]

Sample Output 2:

[1]

```python
# Import the literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
n = int(input())
arr = literal_eval(input())

def product_except_self(arr):
    # Write your code here

    # If the list contains only one element, return [1]
    length = len(arr)
    if length == 1:
        return [1]

    # Validation (optional)
    if length < 1 or length > 10**3:
        return "Invalid input"

    # Initialize prefix and suffix products
    prefix_products = [1] * length
    suffix_products = [1] * length
    output_list = [1] * length

    # Build prefix_products
    for i in range(1, length):
        prefix_products[i] = prefix_products[i-1] * arr[i-1]

    # Build suffix_products
    for i in range(length-2, -1, -1):
        suffix_products[i] = suffix_products[i+1] * arr[i+1]

    # Calculate result by multiplying prefix and suffix products
    for i in range(length):
        output_list[i] = prefix_products[i] * suffix_products[i]

    return output_list

# Print the output
print(product_except_self(arr))
```

Write a function that takes a list of integers and returns a new list containing all elements that appear exactly twice in the list.
