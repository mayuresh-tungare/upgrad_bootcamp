String Compression

Write a program to perform basic string compression using the counts of repeated characters. For each group of consecutive repeating characters in the input string, replace it with the character followed by the number of repetitions. If the compressed string is not shorter than the original string, return the original string.





Input Format:

A single string 's' consisting of only lowercase English letters.



Output Format:

A single string: the compressed version of 's' if it is shorter than 's'; otherwise, 's'.



Constraints:

1 <= len(s) <= 1000

's' contains only lowercase English alphabets [a-z]



Sample Input 1:

aabcccccaaa



Sample Output 1:

a2b1c5a3



Sample Input 2:

abcd



Sample Output 2:

abcd

```python
# Taking Input
s = input()

# Write your code here
def compress(s):
    if not s:
        return s
    else:
        compressed = []
        count = 1

        for i in range (1, len(s)):
            if s[i] == s[i-1]:
                count = count + 1
            else:
                compressed.append(s[i-1])
                compressed.append(str(count))
                count = 1

    compressed.append(s[-1])
    compressed.append(str(count))
    compressed_string = ''.join(compressed)
    return compressed_string if len(compressed_string) < len(s) else s

# Print the Output
print(compress(s))
```

Check if a String is a Palindrome
Write a Python program that checks whether a given string is a palindrome.



A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward.



The program must ignore the following when checking for a palindrome:

Spaces
Punctuation (like commas, periods, exclamation marks)
Case (uppercase vs. lowercase letters)


Input Format:

A string 's' that may contain letters, spaces, and punctuation.



Output Format:

True if the string is a palindrome (ignoring spaces, punctuation, and case).
False otherwise.


Constraints:

The input string contains printable characters (letters, spaces, punctuation).



Sample Input :

"madam"



Sample Output:

True

```python
# Taking input
s=input()

def is_palindrome(s):
    # Write your code here

    if not s:
        return s

    clear_string = "".join(char.lower() for char in s if char.isalnum())
    return clear_string == clear_string[::-1]

# Print the output
print(is_palindrome(s))
```

Write a function ‘are_anagrams(str1, str2)’ that takes two strings ‘str1’ and ‘str2’ as input and returns True if they are anagrams of each other, and False otherwise. The function should be case-insensitive and should ignore spaces between words.



An anagram is when two strings contain the same characters, but possibly in a different order.




Input Format:

Two strings, ‘str1’ and ‘str2’, which may contain spaces and are case-insensitive.



Output Format:

True if the two strings are anagrams, and False otherwise.



Constraints:

The input strings can be of length between 0 and 100.
The input strings can contain alphabetic characters (uppercase and lowercase) and spaces.




Sample input:

silent



Sample Output:

```python
# Taking input
str1 = input()
str2 = input()

def are_anagrams(str1, str2):
    # Write your code here

    new_str1 = str1.replace(" ","").lower()
    new_str2 = str2.replace(" ","").lower()

    return sorted(new_str1) == sorted(new_str2)

# Print the result
print(are_anagrams(str1, str2))
```

Biggest Product from Two Numbers

Write a function ‘max_product(lst)’ that takes a list ‘lst’ of integers and returns the maximum product of any two distinct elements from the list. The list can contain both positive and negative integers.



Input Format:

A list of integers.



Output Format:

The maximum product of any two distinct elements from the list.



Constraints:

The list will contain at least two integers.
The integers in the list may be positive, negative, or zero.


Sample Input:

[1, 10, -5, 1]



Sample Output:

10

```python
# Import literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
lst = literal_eval(input())

def max_product(lst):
    # Write your code here

    sorted_list = sorted(lst) #sorting the given list

    product_min = sorted_list[0] * sorted_list[1]
    product_max = sorted_list[-1] * sorted_list[-2]

    if product_max >= product_min:
        return product_max
    else:
        return product_min

# Print the output
print(max_product(lst))
```

You are given two lists of integers; 'list1' and 'list2'. Merge these lists into one sorted list and return the median value.



If the total number of elements is odd, return the middle element.
If even, return the average of the two middle elements.


This task requires handling both odd and even-length combined lists.



Input Format:

Two lines:

A list of integers (list1)
Another list of integers (list2)


Output Format:

A single float or integer representing the median



Constraints:

1 ≤ len(list1) + len(list2) ≤ 1000
Elements in the list can be positive, negative, or zero


Sample Input:

[1, 3, 5, 7, 9]

[2, 4, 6, 8, 10]



Sample Output:

5.5

```python
# Import literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
list1 = literal_eval(input())
list2 = literal_eval(input())

def find_median_lists(list1, list2):
    # Write your code here

    new_list = sorted(list1 + list2) #merge and sort two list
    n = len(new_list)
    mid = n//2

    if n % 2 == 1:
        return new_list[mid]
    else:
        return (new_list[mid-1] + new_list[mid])/2

# Print the output
print(find_median_lists(list1, list2))
```

Write a program that defines a function 'count_above_average(numbers)' which returns the count of elements in the list that are strictly greater than the average of all elements.



Input Format:

A single line containing a list of integers.



Output Format:

A single integer that represents the count of elements strictly greater than the average of the list.



Constraints:

The list contains at least 1 and at most 100 integers.
Each integer lies between -10⁵ and 10⁵.


Sample Input 1:

[1, 2, 3, 4, 5]



Sample Output 1:

2



Explanation:

The average of the list is 3. Elements greater than 3 are 4 and 5, so the output is 2.





Sample Input 2:

[10, 10, 10, 10]



Sample Output 2:

0

```python
# Import literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
numbers = literal_eval(input())

def count_above_average(numbers):
    # Write your code here

    # calculate average and initiate count = 0
    avg = int(sum(numbers)/len(numbers))
    count = 0

    for i in range(len(numbers)):
        if numbers[i] > avg:
            count = count + 1
        else:
            continue

    return count

# Print the output
print(count_above_average(numbers))
```

Remove Duplicates from Sorted List

Write a function ‘remove_duplicates(lst)’ that removes duplicates from a sorted list. The function should modify the list in-place and return the length of the list after removing duplicates.



Input Format:

A sorted list of integers 'lst'.



Output Format:

An integer representing the new length of the list after removing duplicates.



Constraints:

The input list will be sorted.

The list will contain at least 0 elements.





Sample Input:

[1, 1, 2, 2, 3]



Sample Output:

 3

```python
# Import literal_eval library to safely evaluate string input as a Python
from ast import literal_eval

# Taking input
lst = literal_eval(input())

def remove_duplicates(lst):
    # Write your code here

    new_lst = []
    for i in range(len(lst)):
        if lst[i] in new_lst:
            continue
        else:
            new_lst.append(lst[i])

    return len(new_lst)

# Print the output
print(remove_duplicates(lst))
```

Write a function ‘is_balanced(s)’ that checks if the parentheses in the given string are balanced. The function should also check for balanced brackets [] and curly braces {} in addition to the regular parentheses ().



A string is considered balanced if:

Every opening bracket has a corresponding closing bracket.
The brackets are closed in the correct order.


Input Format:

A single string ‘s’ consisting of characters such as (, ), {, }, [, ], and other non-bracket characters.



Output Format:

True if the string has balanced brackets and False otherwise.



Constraints:

 1<=len(s)<=105
The string can contain any combination of the characters (), {}, [], and other non-bracket characters.




Sample Input:

{[()]}



Sample Output:

True

```python
s = input()

def is_balanced(s):
    # Write your code here

    stack = []
    # Mapping of closing brackets to their corresponding opening brackets
    bracket_map = {')': '(', '}': '{', ']': '['}

    for char in s:
        # If it's an opening bracket, push to stack
        if char in bracket_map.values():
            stack.append(char)
        # If it's a closing bracket
        elif char in bracket_map:
            # If stack is empty or top of stack doesn't match, not balanced
            if not stack or stack[-1] != bracket_map[char]:
                return False
            stack.pop()
        # Ignore non-bracket characters

    # If stack is empty, all brackets matched
    return len(stack) == 0


# Print the output
print(is_balanced(s))
```

Mask Bank Account Number

Write a function 'mask_account(account)' that takes a bank account number as a string.



The function should return a masked version of the account number where:

The first four digits remain unchanged.
The last four digits remain unchanged.
All characters between the first four and last four digits are replaced with the asterisk character (*).


Input Format:

A single string account representing a bank account number.



Output Format:

A string representing the masked bank account number, where the characters in the middle are masked except for the first and last four digits.



Constraints:

The bank account number is valid and contains at least 8 characters.
The bank account number consists only of digits.


Sample Input:

1234567890123456



Sample Output:

1234********3456

```python
# Taking input
account = input()

def mask_account(account):
    # Write your code here

    if account.isdigit():
        if len(account) >= 8:
            masked=""
            for i in range(len(account)):
                if i < 4 or ((len(account)-i) <= 4):
                    masked = masked + account[i]
                else:
                    masked = masked + '*'
            return masked
        else:
            return "Invalid account number"

# Print the output
print(mask_account(account))
```

Check Isomorphic Strings

Two strings 's' and 't' are isomorphic if the characters in 's' can be replaced to get 't'.



All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.



Input Format:

The program expects two strings as input.



Output Format:

True if two strings are isomorphic, False otherwise.



Constraints:

1 <= len(s)<= 5 * 104
len(t) == len(s)
's' and 't' consist of any valid ASCII character.




Sample Input 1:

"egg "

"add"



Sample Output 1:

True



Sample Input 2:

"foo"

"bar"



Sample Output 2:

False

```python
# Taking input
s = input()
t = input()

def are_isomorphic(s, t):
    # Write your code here

    if len(s) >= 1 and len(s) <= 5 * 10 ** 4:
        if len(t) >= 1 and len(s) <= 5 * 10 ** 4:
            if len(s) == len(t):
                mapping_s_t = {}
                mapped_t_chars = set()
                for char_s, char_t in zip(s, t):
                    if char_s in mapping_s_t:

                        if mapping_s_t[char_s] != char_t:
                            return False
                        else:
                            # If char_t is already mapped by another char_s, return False
                            if char_t in mapped_t_chars:
                                return False

                                # Establish new mapping
                                mapping_s_t[char_s] = char_t
                                mapped_t_chars.add(char_t)
            else:
                return False
    return True
```

Check String Rotation

You are given two strings, 's1' and 's2'. Write a program to check if 's2' is a rotation of 's1'.



Two strings are said to be rotations of each other if they contain the same characters in the same order but starting from a different position and wrapping around. For example, "erbottlewat" is a rotation of "waterbottle" because the characters from the start have been shifted to the end without changing their order.



Input Format:

Two lines of input:

First line: string 's1'  
Second line: string 's2'


Output Format:

Print "Yes" if 's2' is a rotation of 's1', otherwise print "No"



Constraints:

1 <= len(s1), len(s2) <= 1000  

Both strings contain only lowercase english letters [a-z]



Sample Input 1:  

waterbottle  

erbottlewat  



Sample Output 1:  

Yes  



Sample Input 2:  

hello

lehlo  



Sample Output 2:

No

```python
# Taking Input
s1 = input()
s2 = input()

# Write your code here
def is_rotation(s1, s2):

    # Perform basic validations
    if len(s1)<=1 or len(s2) > 1000:
        return False
    if len(s2)<=1 or len(s1) > 1000:
        return False
    if len(s1)!=len(s2):
        return False

    # Concatenate s1 with itself, and check if s2 is a substring
    return s2 in (s1 + s1)

# Print the Output
if is_rotation(s1, s2):
    print("Yes")
else:
    print("No")
```
