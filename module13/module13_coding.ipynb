Calculate nth Fibonacci Using Recursion

You are given an integer n in the range 0 to 30. Your task is to write a recursive function fibonacci(n) that returns the Fibonacci number at the nth position in the sequence. The Fibonacci sequence starts with 0 at position 0 and 1 at position 1. Each subsequent number is the sum of the two preceding numbers in the sequence.


Input Format



A single value (int) n representing the position in the Fibonacci sequence

Output Format


A single integer representing the Fibonacci number at position n

Constraints


The input integer n must be in the range 0 to 30, inclusive

Example Case 1


Input


0


Output


0



Example Case 2


Input


3


Output


2

```python
def fibonacci(n):
    # Code here

    if n == 0:
        return 0
    elif n == 1:
        return 1
    elif n > 1:
        return (fibonacci(n-1) + fibonacci(n-2))
    else:
        return "Invalid"

# (do not edit)
print(fibonacci(int(input())))
```

Fruit Basket Probability

You are given a list of dictionaries, where each dictionary represents a fruit in a basket. Each dictionary contains the keys fruit and count. The key fruit is a string representing the type of fruit (e.g., 'Apple', 'Banana', 'Orange', etc.), and count is an integer representing the number of that fruit in the basket.



Write a Python function fruit_basket_probability(basket_data) that calculates the probability of selecting a specific type of fruit (e.g., 'Apple') from the basket. The function should return the probability as a decimal value, which is the ratio of the count of the target_fruit to the total number of fruits in the basket.

Input Format

A single tuple as input, containing two elements:


A list of dictionaries, where each dictionary contains:


'fruit' (str): The name of the fruit (e.g., 'Apple', 'Banana', 'Orange', etc.)
'count' (int): The number of that fruit in the basket

A string target_fruit representing the type of fruit to calculate the probability for


Output Format

A decimal value representing the probability of selecting the target_fruit
If the target_fruit is not found in the basket, its probability is 0.0
If the total number of fruits in the basket is zero, the probability of selecting any fruit is 0.0

Constraints

Probability should be rounded to two decimal places

Example Case 1

Input

([{'fruit': 'apple', 'count': 3}, {'fruit': 'banana', 'count': 2}, {'fruit': 'orange', 'count': 5}], 'apple')

Output

0.3

Example Case 2

Input

([{'fruit': 'apple', 'count': 0}, {'fruit': 'banana', 'count': 0}, {'fruit': 'orange', 'count': 0}], 'apple')

Output

0.0

```python
from ast import literal_eval

def fruit_basket_probability(basket_data):
    # Code here
    basket, fruit = basket_data
    fruit_count = 0
    total_count = 0

    for item in basket:
        total_count = total_count + item['count']
        if item['fruit'] == fruit:
            fruit_count = fruit_count + item['count']

    if total_count != 0:
        return (fruit_count/total_count)
    else:
        return 0.0

# (do not edit)
print(round(fruit_basket_probability(literal_eval(input())), 2))
```

Sorted Frequency Encoding

You are given a string containing both uppercase and lowercase English letters. Your task is to define a Python function sorted_frequency_encoding(s) that accepts a string and returns a new string in which characters are sorted in lexicographical order (i.e., all uppercase letters come before lowercase ones), each unique character appears only once, and is immediately followed by the number of times it occurs in the input.


Input Format

A single string (str) consisting of uppercase and lowercase English letters

Output Format

A single string (str) where each unique character from the input appears once, followed by its frequency

Constraints

You may use the built-in sorted() function to obtain characters in lexicographical order
All characters in the input are alphabetic (no digits, symbols, or whitespace)

Example Case 1


Input


aabBcC


Output



Ab2B1C1c1


Example Case 2


Input


AAAaaa


Output


A3a3

```python
def sorted_frequency_encoding(s):
    # Code here
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1

    uppers = [c for c in freq if c.isupper()]
    lowers = [c for c in freq if c.islower()]
    uppers.sort()
    lowers.sort()
    chars = uppers + lowers

    result = ''
    for c in chars:
        result += c + str(freq[c])
    return result

# (do not edit)
print(sorted_frequency_encoding(input()))
```

Employee Salary Update

Write a Python function update_employee_salaries(employee_data) that takes a dictionary of employee names and their respective salaries and returns a new dictionary where each employee's salary is updated by applying a 10% increase. The function should take the original dictionary as input and return the updated dictionary.


Input Format

A dictionary where the keys are employee names (str), and the values are their current salaries (int or floats)

Output Format

A dictionary where each key is the employee's name, and the value is the updated salary, increased by 10%

Constraints

The updated salary must be rounded down to the nearest integer using integer truncation (i.e., using int())

Example Case 1

Input

{'Ravi': 50000, 'Anjali': 60000, 'Amit': 55000}


Output

{'Ravi': 55000, 'Anjali': 66000, 'Amit': 60500}



Example Case 2


Input


{'Manoj': 45000.75, 'Sneha': 32000.50, 'Rakesh': 54000.25}


Output


{'Manoj': 49500, 'Sneha': 35200, 'Rakesh': 59400}

```python
from ast import literal_eval

def update_employee_salaries(employee_data):
     # Code here

     employee_data_new = employee_data
     for employee in employee_data_new:
          employee_data_new[employee] = int(employee_data_new[employee] + employee_data_new[employee] * 0.1)
     return employee_data_new

# (do not edit)
print(update_employee_salaries(literal_eval(input())))
```

Team Selection Combinations

Write a Python function team_selection_combinations(employees , k) that takes a list of employee names and determines how many ways you can select a team of k employees from a pool of n employees, where the order of selection does not matter. You should calculate the number of possible combinations for selecting the team.


Input Format

A single tuple as input, containing two elements:


A list of strings employees, where each string is an employee's name
An integer k representing the number of employees to select for the team

Output Format

An integer representing the total number of ways to select k employees from the n employees in the list

Constraints

N/A

Example Case 1


Input


(['Amit', 'Rohit', 'Priya', 'Anjali', 'Neha'],3)


Output

10


Example Case 2


Input



(['Kiran'], 1)


Output


1

```python
from ast import literal_eval
import itertools

def team_selection_combinations(employees , k):
    # Code here
    ways = []
    ways = list(itertools.combinations(employees,k))
    return len(ways)

# (do not edit)
print(team_selection_combinations(*literal_eval(input())))
```
