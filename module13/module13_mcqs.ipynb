A survey finds that 60% of participants enjoy reading fiction, 45% enjoy reading non-fiction, and 25% enjoy both. If a participant does not enjoy fiction, what is the probability that they enjoy non-fiction?

ANSWER:
```python
# Given probabilities
p_f = 0.60
p_nf = 0.45
p_f_and_nf = 0.25

# Probability of not enjoying fiction
p_not_f = 1 - p_f

# Probability of enjoying non-fiction but not fiction
# P(NF and not F) = P(NF) - P(NF and F)
p_nf_and_not_f = p_nf - p_f_and_nf

# Conditional probability: P(NF | not F)
p_nf_given_not_f = p_nf_and_not_f / p_not_f

print(f"{p_not_f=}")
print(f"{p_nf_and_not_f=}")
print(f"{p_nf_given_not_f=}")



```
