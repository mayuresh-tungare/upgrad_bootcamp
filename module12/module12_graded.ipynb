**Rate of Change in Retail Revenue**


A retail company models its daily revenue using a polynomial function that depends on the number of days in a given month. This model helps the finance team predict how revenue is expected to change throughout the month.



For each month, the company uses a different polynomial function that represents revenue as a function of time (in days). The revenue on any day t  is given by:



R(t)=a0+a1t+a2t2+.......+antn



Your task is to define the function marginal_revenue(data) to compute the marginal revenue—that is, the rate of change of revenue with respect to time—on a specific day.



This is done by calculating the first derivative of the polynomial and evaluating it at the given day



Input Format



A tuple containing two elements:

A tuple of elements (int) representing the coefficients of the polynomial function in decreasing order of powers ( for example, (3, -5, 20) represents R(t) = 3t2 - 5t + 20 )
A value (int) representing the specific day of the month on which the marginal revenue is to be calculated


Output Format



A value (int) representing the rate of change of revenue (marginal revenue) on the given day



Constraints

N/A


Example Case 1



Input

((200, 100, 200), 3)  



Output

1300



Example Case 2  


Input


((500, -30, 200), 4)  



Output


3970

```python
from ast import literal_eval

def marginal_revenue(data):
    # Code here
    coeffs, t = data
    n = len(coeffs)
    marg_rev = 0

    # Iterate through coefficients to calculate the derivative
    for i, coeff in enumerate(coeffs):
        # Power is determined by the position in the tuple (decreasing order)
        power = n - 1 - i

        # Power rule: d/dt(a * t^n) = n * a * t^(n-1)
        if power > 0:
            derivative = power * coeff * (t ** (power - 1))
            marg_rev += derivative

    return marg_rev

# (do not edit)
print(marginal_revenue(literal_eval(input())))
```

Calculate Grant Combinations for Funding

You are a program manager at a national research funding agency. Each year, the agency receives numerous grant proposals across various scientific disciplines. To ensure fair and effective allocation of limited funds, the agency has strict rules about how many grants can be awarded within certain research areas and how many can be awarded to new vs. established principal investigators (PIs).


Your task is to write a python function calculate_grant_combinations(total_proposals, new_pi_proposals, established_pi_proposals, max_new_pi_grants, total_grants_to_award) that calculates the number of unique ways to award a fixed number of grants, adhering to the following rules:


Only a limited number of grants can be awarded to new PIs
The rest must go to established PIs
No grant can be awarded more than once, and each PI can receive at most one grant

Input Format


A tuple containing the following five elements:


total_proposals: An integer representing the total number of unique research proposals received
new_pi_proposals: An integer representing the number of proposals from new (early-career) Principal Investigators
established_pi_proposals: An integer representing the number of proposals from established (senior) Principal Investigators
max_new_pi_grants: An integer representing the maximum number of grants that can be awarded to new PIs
total_grants_to_award: An integer representing the total number of grants the agency plans to award

Output Format


Return the total number of unique ways the agency can select grants under the given constraints

Constraints

N/A


Example Case 1


Input


(10, 4, 6, 2, 4)

Output

185

Example Case 2

Input

(10, 5, 5, 5, 5)

Output

252

```python
from ast import literal_eval
import itertools
import math

def calculate_grant_combinations(total_proposals, new_pi_proposals, established_pi_proposals, max_new_pi_grants, total_grants_to_award):
    # Code here
    # At the outset, total ways are 0 (uncounted so far!)
    total_ways = 0

    # Iterate through all possible number of grants for New PIs
    # From 0 up to the maximum allowed (max_new_pi_grants)
    for k_new in range(max_new_pi_grants + 1):

        # Calculate how many established PIs must get grants in this scenario
        k_est = total_grants_to_award - k_new

        # Validation checks
        if k_new <= new_pi_proposals and k_est <= established_pi_proposals and k_est >= 0:
            # Ways to pick New PIs * Ways to pick Established PIs
            ways = math.comb(new_pi_proposals, k_new) * math.comb(established_pi_proposals, k_est)
            total_ways += ways

    return total_ways

# (do not edit)
print(calculate_grant_combinations(*literal_eval(input())))
```
